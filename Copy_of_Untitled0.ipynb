```python
#Question 1

print(round(3.6))            # Rounds 3.6 to the nearest whole number
print(round(3.14159, 2))     # Rounds Ï€ to two decimal places
print(min(4, 2, 8, 1))       # Finds the smallest value among the given numbers
print(ord('A'))              # Returns the ASCII code for the character 'A'
print(chr(65))               # Converts ASCII code 65 back to its character representation
#Question 2

import math
print(math.sqrt(25))         # Computes the square root of 25
print(math.factorial(5))     # Finds the factorial of 5

#Question 3

def my_abs(num):
    if num < 0:
        print(-num)          # If negative, print its positive value
    else:
        print(num)           # Otherwise, print the number itself

my_abs(-10)
my_abs(7)
#Question 4

def my_abs(num):
    if num < 0:
        return -num          # Returns absolute value for negative numbers
    else:
        return num           # Returns the number unchanged

print(my_abs(-12))
print(my_abs(8))

#Question 5

def my_abs(num):
    # Ensure only numeric input is processed
    if not isinstance(num, (int, float)):
        return "Error: Please provide a number."
    if num < 0:
        return -num
    else:
        return num

print(my_abs(-5))
print(my_abs(3.2))
print(my_abs("hello"))

#Question 6

def quadratic(a, b, c):
    # The value of 'a' must be non-zero to form a quadratic equation
    if a == 0:
        return "Error: For a quadratic equation, 'a' must not be zero."

    d = b**2 - 4*a*c   # Calculate discriminant

    if d > 0:          # Two real, distinct solutions
        r1 = (-b + math.sqrt(d)) / (2 * a)
        r2 = (-b - math.sqrt(d)) / (2 * a)
        return r1, r2

    elif d == 0:       # One real repeated solution
        r = -b / (2 * a)
        return r, r

    else:              # Complex solutions
        real = -b / (2 * a)
        imag = math.sqrt(-d) / (2 * a)
        return complex(real, imag), complex(real, -imag)

print(quadratic(1, -3, 2))
print(quadratic(1, 2, 1))
print(quadratic(1, 1, 1))
```
