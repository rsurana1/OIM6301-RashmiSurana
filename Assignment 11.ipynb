```python
# -------------------------
# STRING BASICS
# -------------------------

team = "New England Patriots"

print(team[0])
print(len(team))
print(team[-1])

# -------------------------
# TRAVERSAL
# -------------------------

for letter in team:
    print(letter)

# -------------------------
# EXERCISE 01
# -------------------------

prefixes = "JKLMNOPQ"
suffix = "ack"

for letter in prefixes:
    if letter == "O" or letter == "Q":
        print(letter + "u" + suffix)
    else:
        print(letter + suffix)

# -------------------------
# STRING SLICING
# -------------------------

print(team[0:11])
print(team[12:20])
print(team[:11])
print(team[12:])
print(team[:])
print(team[::2])

# -------------------------
# IMMUTABILITY
# -------------------------

new_team = team[:12] + "Seahawks"
print(new_team)
print(team)

# -------------------------
# SEARCH FUNCTION
# -------------------------

def find(word, letter):
    index = 0
    while index < len(word):
        if word[index] == letter:
            return index
        index += 1
    return -1

print(find(team, "E"))

# -------------------------
# EXERCISE 02
# -------------------------

def count(word, letter):
    total = 0
    for c in word:
        if c == letter:
            total += 1
    return total

print(count(team, "a"))

# -------------------------
# STRING METHODS
# -------------------------

print(team.upper())
print(team.find("a"))
print(team.find("En"))
print(team.find("a", 10))

name = "bob"
print(name.find("b", 1, 2))

# -------------------------
# IN OPERATOR
# -------------------------

print("a" in team)
print("Boston" in team)

def in_both(word1, word2):
    for c in word1:
        if c in word2:
            print(c)

in_both("hello", "world")

# -------------------------
# EXERCISE 04
# -------------------------

def any_lowercase_1(s):
    for c in s:
        if c.islower():
            return True
        else:
            return False

def any_lowercase_2(s):
    for c in s:
        if "c".islower():
            return "True"
        else:
            return "False"

def any_lowercase_3(s):
    for c in s:
        flag = c.islower()
    return flag

def any_lowercase_4(s):
    flag = False
    for c in s:
        flag = flag or c.islower()
    return flag

def any_lowercase_5(s):
    for c in s:
        if not c.islower():
            return False
    return True

# -------------------------
# EXERCISE 05
# -------------------------

def rotate_word(word, shift):
    result = ""
    for c in word:
        if "a" <= c <= "z":
            result += chr((ord(c) - ord("a") + shift) % 26 + ord("a"))
        elif "A" <= c <= "Z":
            result += chr((ord(c) - ord("A") + shift) % 26 + ord("A"))
        else:
            result += c
    return result

print(rotate_word("cheer", 7))
print(rotate_word("melon", -10))
print(rotate_word("IBM", -1))
```
