```python
import time

current = time.time()

days = int(current // 86400)
remaining = current % 86400

hours = int(remaining // 3600)
remaining %= 3600

minutes = int(remaining // 60)
seconds = int(remaining % 60)

print(f"Days since epoch: {days}")
print(f"Time of day: {hours:02d}:{minutes:02d}:{seconds:02d}")
```

```python
a = int(input("Enter first integer: "))
b = int(input("Enter second integer: "))

print(f"{a} + {b} = {a + b:6}")
print(f"{a} - {b} = {a - b:6}")
print(f"{a} * {b} = {a * b:6}")
print(f"{a} / {b} = {a / b}")
print(f"{a} ^ {b} = {a ** b:6}")
```
