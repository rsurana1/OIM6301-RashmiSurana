```python
# Iteration Assignment - All Exercises in One File

import math

# =========================
# Exercise 01 - Using for loops
# =========================

def exercise1_for():
    print("=== Exercise 01 - for loops ===")

    # 1. Sum of integers from 1 to 10
    sum_1_10 = 0
    for i in range(1, 11):
        sum_1_10 += i
    print("Sum 1 to 10 (for):", sum_1_10)

    # 2. Sum of integers from 1 to 1000
    sum_1_1000 = 0
    for i in range(1, 1001):
        sum_1_1000 += i
    print("Sum 1 to 1000 (for):", sum_1_1000)

    # 3. Sum of all odd numbers from 1 to 1000
    sum_odd_1_1000 = 0
    for i in range(1, 1001, 2):
        sum_odd_1_1000 += i
    print("Sum of odd numbers 1 to 1000 (for):", sum_odd_1_1000)

    # 4. Sum of all even numbers from 1 to 1000
    sum_even_1_1000 = 0
    for i in range(2, 1001, 2):
        sum_even_1_1000 += i
    print("Sum of even numbers 1 to 1000 (for):", sum_even_1_1000)
    print()


# =========================
# Exercise 01 - Using while loops
# =========================

def exercise1_while():
    print("=== Exercise 01 - while loops ===")

    # 1. Sum of integers from 1 to 10
    total = 0
    i = 1
    while i <= 10:
        total += i
        i += 1
    print("Sum 1 to 10 (while):", total)

    # 2. Sum of integers from 1 to 1000
    total = 0
    i = 1
    while i <= 1000:
        total += i
        i += 1
    print("Sum 1 to 1000 (while):", total)

    # 3. Sum of all odd numbers from 1 to 1000
    total = 0
    i = 1
    while i <= 1000:
        total += i
        i += 2
    print("Sum of odd numbers 1 to 1000 (while):", total)

    # 4. Sum of all even numbers from 1 to 1000
    total = 0
    i = 2
    while i <= 1000:
        total += i
        i += 2
    print("Sum of even numbers 1 to 1000 (while):", total)
    print()


# =========================
# Exercise 02 - Loop behavior questions
# =========================

def exercise2():
    print("=== Exercise 02 - Loop behavior ===")

    # -------- Part 1 --------
    print("--- Part 1 ---")
    iteration = 0
    count = 0
    while iteration < 5:
        # the variable 'letter' in the loop stands for every
        # character, including spaces and commas!
        for letter in "hello, world":
            count += 1
        print(f"Iteration {iteration}; count is: {count}")
        iteration += 1

    print()

    # -------- Part 2 --------
    print("--- Part 2 ---")
    iteration = 0
    while iteration < 5:
        count = 0
        for letter in "hello, world":
            count += 1
        print(f"Iteration {iteration}; count is: {count}")
        iteration += 1

    print()

    # -------- Part 3 --------
    print("--- Part 3 ---")
    iteration = 0
    while iteration < 5:
        count = 0
        for letter in "hello, world":
            count += 1
            if iteration % 2 == 0:
                break
        print(f"Iteration {iteration}; count is: {count}")
        iteration += 1

    print()


# =========================
# Break question (mysum)
# =========================

def exercise_break():
    print("=== Break question ===")
    mysum = 0
    for i in range(5, 11, 2):
        mysum += i
        if mysum == 5:
            break
    print("Final value of mysum:", mysum)
    print()


# =========================
# Exercise 03 - mysqrt and test_square_root (Newton's method)
# =========================

def mysqrt(a):
    """
    Use Newton's method to approximate sqrt(a).
    """
    x = a / 2  # initial guess
    epsilon = 1e-10

    while True:
        y = (x + a/x) / 2
        if abs(y - x) < epsilon:
            return y
        x = y


def test_square_root():
    print("=== Exercise 03 - test_square_root ===")
    print("a   mysqrt(a)           math.sqrt(a)        diff")
    print("-   ---------           ------------        ----")

    for a in range(1, 10):
        my_val = mysqrt(a)
        math_val = math.sqrt(a)
        diff = abs(my_val - math_val)
        print(f"{a:<3} {my_val:<19} {math_val:<19} {diff}")
    print()


# =========================
# Run all exercises
# =========================

# In a .py file, this will run when you execute the script.
# In Google Colab, you can just run this whole cell.

exercise1_for()
exercise1_while()
exercise2()
exercise_break()
test_square_root()
```
