```python
# Exercise 1
def check_fermat(a, b, c, n):
    if n > 2 and (a**n + b**n == c**n):
        print("Holy smokes, Fermat was wrong!")
    else:
        print("No, that doesn't work.")


def prompt_fermat_check():
    a = int(input("Enter value for a: "))
    b = int(input("Enter value for b: "))
    c = int(input("Enter value for c: "))
    n = int(input("Enter value for n: "))

    check_fermat(a, b, c, n)


# Example (uncomment to run)
prompt_fermat_check()

def calculate_bmi(weight, height):
    bmi = weight / (height ** 2)
    return bmi


def get_bmi_category():
    weight = float(input("Enter your weight in kilograms: "))
    height = float(input("Enter your height in meters: "))

    bmi = calculate_bmi(weight, height)

    if bmi < 18.5:
        category = "Underweight"
    elif bmi < 25:
        category = "Normal weight"
    elif bmi < 30:
        category = "Overweight"
    else:
        category = "Obesity"

    print(f"Your BMI is {bmi:.2f}, Category: {category}")
    return category


# Example (uncomment to run)
get_bmi_category()

# Exercise 2
def factorial(n):
    if n < 0:
        return "Error: Factorial is not defined for negative numbers."
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)


# Example
n = int(input("Enter an integer to compute factorial: "))
print(f"Factorial of {n} is {factorial(n)}")

def fibonacci(n):
    if n < 0:
        return "Error: Fibonacci is not defined for negative numbers."
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)


# Example
n = int(input("Enter an integer to find Fibonacci number: "))
print(f"Fibonacci number at position {n} is {fibonacci(n)}")

def gcd(a, b):
    if b == 0:
        return a
    return gcd(b, a % b)


# Example
x = int(input("Enter first positive integer: "))
y = int(input("Enter second positive integer: "))
print(f"GCD of {x} and {y} is {gcd(x, y)}")

def tower_of_hanoi(n, source, target, auxiliary):
    if n == 1:
        print(f"Move disk 1 from {source} to {target}")
        return
    tower_of_hanoi(n-1, source, auxiliary, target)
    print(f"Move disk {n} from {source} to {target}")
    tower_of_hanoi(n-1, auxiliary, target, source)


# Example
n = int(input("Enter number of disks for Tower of Hanoi: "))
tower_of_hanoi(n, 'A', 'C', 'B')
```
